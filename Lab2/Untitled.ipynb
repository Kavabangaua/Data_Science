```
import pandas as pd
import numpy as np

data_np = np.random.randint(0, 100, size=(5, 3))
df_from_array = pd.DataFrame(data_np, columns=['A', 'B', 'C'])

#Створюємо DataFrame за допомогою словника
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 32, 23, 45, 28],
    'Score': [85, 90, 78, 88, 92]
}
df_from_dict = pd.DataFrame(data_dict)

#Відображаємо перші 3 рядки DataFrame
head_data = df_from_dict.head(3)

description = df_from_dict.describe()

#Індексація за допомогою iloc (по позиції)
iloc_data = df_from_dict.iloc[1:4]

#Індексація за допомогою loc (по мітці)
loc_data = df_from_dict.loc[1:3, ['Name', 'Score']]

# Виводимо результати
print("DataFrame з NumPy масиву:\n", df_from_array)
print("\nDataFrame зі словника:\n", df_from_dict)
print("\nПерші 3 рядки:\n", head_data)
print("\nОпис статистики:\n", description)
print("\nІндексація з iloc:\n", iloc_data)
print("\nІндексація з loc:\n", loc_data)
```

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error

#Генеруємо дані для лінійної функції y = 2x + 1
x = np.linspace(0, 10, 50)
y_true = 2 * x + 1

#Створюємо випадкові помилки для даних (помилки вимірювання)
errors = np.random.normal(0, 1, size=x.shape)
y_noisy = y_true + errors

# Plot: Будуємо графік
plt.figure(figsize=(8, 6))
plt.plot(x, y_true, label='Лінійна функція (y = 2x + 1)', color='blue')
plt.scatter(x, y_noisy, label='Дані з помилками', color='red')
plt.title('Графік лінійної функції та даних з помилками')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

#Обчислюємо метрики похибок
mae = mean_absolute_error(y_true, y_noisy)
mse = mean_squared_error(y_true, y_noisy)

#Записуємо результати у CSV файл
results = pd.DataFrame({
    'MAE': [mae],
    'MSE': [mse]
})
results.to_csv('error_metrics.csv', index=False)

# Виводимо обчислені метрики
print(f'MAE: {mae}')
print(f'MSE: {mse}')
```
